```python
import networkx as nx
import pandas as pd
```

```python
import glob
all_files = glob.glob('/home/data/sdb/wt/MetGraph/EnzGraphs/*.tsv')
```

```python
###read all graph
from tqdm import tqdm
all_graph = []
for i in tqdm(range(0, len(all_files))):
    tmp = pd.read_csv(all_files[i],sep="\t")
    g = nx.from_pandas_edgelist(tmp, source="from", target="to")
    g = nx.convert_node_labels_to_integers(g)
    all_graph.append(g)
```

```python
from karateclub import Graph2Vec
```

```python
len(all_graph)
```

```python
model = Graph2Vec()
model.fit(all_graph)
X = model.get_embedding()
```

```python
X.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(X, X)
```

```python
sim.shape,sim
```
